```python

def convert_to_float(x):
  try:
    x = float(x)
    if x <= 0:
      x = None
  except:
    x = None
  return x

print(df.shape)
df['Shoe size'] = df['Shoe size'].apply(convert_to_float)
#check for None values in the show size column and drop the rows
df.dropna(inplace = True)
print(df.shape)
```

```python
# Step 2: clean-up birthday -> transform to date and check against age
# if age is not right, we update age based on birthday

df['Birthday'] = pd.to_datetime(df['Birthday'], errors='coerce')
df.dropna(inplace=True)
print('Null values in the Birthday column', df['Birthday'].isnull().sum())

# for age we need to take into consideration the month and day too, not just the year
# present day - birthday(year/month/day) = age

# idea: date now - date then = x DAYS -> /365 and get the age
df['Age'] = (pd.to_datetime('today') - df['Birthday']).dt.days // 365

print('Null values in the Age column', df['Age'].isnull().sum())

# check if there are ages = 0 and fix it
df = df[df['Age'] > 0]
df['Age']
df
```

```python
def filterHeight(data):
    data = data.lower()
    if data.isdigit() and int(data) <= 100:
        return int(data)
    if "'" in data:
        data.replace("'", '')
    if '"' in data:
        data.replace('"', '')
    if '~' in data:
        data.replace('~', '')
    if 'feet' in data:
        data.replace('feet', '')
    if 'foot' in data:
        data.replace('foot', '')
    if 'ft' in data:
        data.replace('ft', '')
    if 'inches' in data:
        data.replace('inches', '')
    data.replace(" ", "")

    try:
        result = int(data[0])*12 + int("".join(data[1:]))
        if result < 100:
            return result
        else:
            return None
    except:
        return None
df['Height in inches'] = df['Height in inches'].apply(filterHeight)
df.dropna(inplace=True)
```

```python
df
```

```python
# Step 4: convert this to int
# accounnt for lower/upper case
# extract the unique values in the color column
# we want to assign int representation for them: ENUM
# OR we can use a dict with key and value
# OR we can use a list for values and we use the index
# OR use color representation from encoders
# Result needs to be int or float

# change everything to lower case
df['Favorite Color'] = df['Favorite Color'].str.strip().str.lower()

#find unique values
df['Favorite Color'].unique()

# replace "purple" -> "violet"
df['Favorite Color'] = df['Favorite Color'].replace("purple", "violet")

# replace "pink" -> "red"
df['Favorite Color'] = df['Favorite Color'].replace("pink", "red")

# replace "teal" -> "blue"
df['Favorite Color'] = df['Favorite Color'].replace("teal", "blue")

# replace "lavender" -> "violet"
df['Favorite Color'] = df['Favorite Color'].replace("lavender", "violet")

# replace "any pastel" -> delete it
df['Favorite Color'] = df['Favorite Color'].replace("any pastel", np.nan)

# replace "fire red" -> "red"
df['Favorite Color'] = df['Favorite Color'].replace("fire red", "red")

# replace "amaranth" -> "red"
df['Favorite Color'] = df['Favorite Color'].replace("amaranth", "red")

# replace "deep purple" -> "violet"
df['Favorite Color'] = df['Favorite Color'].replace("deep purple", "violet")

# replace "light blue" -> "blue"
df['Favorite Color'] = df['Favorite Color'].replace("light blue", "blue")

code={'black':1,
      'blue':2.50,
      'cyan':2.17,
      'red':1.65,
      'orange':2,
      'yellow':2.10,
      'violet':2.75,
      'green':2.17,
      'white':3}
df['Favorite Color']=df['Favorite Color'].map(code)


```

```python
df
```

```python
#Step 5: comvert lunch to numerical values
df['Lunch?'].unique()

lunch_val = {
    'Taco bell':-1,
    'Home-cooked':0,
    'Chick-fill-a':1
}

df['Lunch?'] = df['Lunch?'].map(lunch_val)
```

```python
#Step 6: convert games to numerical value
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Favorite Game table-top game' ] = le.fit_transform(df['Favorite Game table-top game' ])
df['Favorite Game table-top game']
```

```python
# Step 7: convert fav. hobby to numerical encoding
# take into account if I have the same hobby in different columns i want it to have the same number
# example if gym = 10, then #1 column should encode gym to 10, #2 should encode gym to 10 also
df['Favorite Hobby #1'] = df['Favorite Hobby #1'].str.strip().str.lower()
df['Favorite Hobby #2'] = df['Favorite Hobby #2'].str.strip().str.lower()
df['Favorite Hobby #3'] = df['Favorite Hobby #3'].str.strip().str.lower()

index = 0
hobby_dict = {}

for el in df['Favorite Hobby #1'].unique():
  if el not in hobby_dict:
    hobby_dict[el] = index
    index+=1

for el in df['Favorite Hobby #2'].unique():
  if el not in hobby_dict:
    hobby_dict[el] = index
    index+=1

for el in df['Favorite Hobby #3'].unique():
  if el not in hobby_dict:
    hobby_dict[el] = index
    index+=1

print(hobby_dict)

# map the values from the dict to the values in the data
df['Favorite Hobby #1'] = df['Favorite Hobby #1'].map(hobby_dict)
df['Favorite Hobby #2'] = df['Favorite Hobby #2'].map(hobby_dict)
df['Favorite Hobby #3'] = df['Favorite Hobby #3'].map(hobby_dict)
```

```python
#convert the timestamp column to seconds taken to fill the for since 6/17/2024
#Example: if a student completed at : 6/`7/2024 15:08:55 -> 55 seconds

#convert timestamp to hour:min:sec
# CONVERT timestamp column to seconds taken to fill form since 6/17/2024 15:08:00

df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# Define the reference time
reference_time = pd.Timestamp('6/17/2024 15:08:55')

# Calculate duration in seconds
df['Duration'] = (df['Timestamp'] - reference_time).dt.total_seconds()

```
